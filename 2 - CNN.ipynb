# CNN

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
```

```python
#Cargar el dataset Fashion Mnist 
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
class_names = ['avión', 'coche', 'pájaro', 'gato', 'ciervo', 'perro', 'rana', 'caballo', 'barco', 'camión']
```

## 1. Preprocesado

```python
#Normalización
X_train = X_train /255.0
X_test = X_test / 255.0
```

## 2. Construcción de la Red Neuronal Convolucional

### 1. Capa convolucional:
 * Filtros: 32 el número de filtros
 * Tamaño del kernel: 3 el tamaño del kernel 3*3
 * padding: same #crea ceros para rellenar y aplicar el filtro (o kernel) a toda la matriz. Valid por el contrarío solo lo hace cuando hay espacio posible, por lo que los bordes pueden no entrar en la operación de convolución
 * Función de Activación: relu
 * input_shape: (32, 32, 3)

### 2. Capa Convolucional y Max Pooling:
 
 * Filtros: 32
 * Tamaño del kernel: 3
 * padding: same
 * Función de Activación: relu

**Hyper parámetros de la capa de MaxPool:**

 * pool_size: 2 #el tamaño es 2*2
 * strides: 2 el salto es de dos en dos
 * padding: valid. Cuando no se puede crear la matriz de max pooling 2*2  no se hace el borde
 
### 3. Capa convolucional
 
 * Filtros: 64
 * Tamaño del kernel: 3
 * padding: same
 * Función de Activación: relu
 
### 4. Capa convolucional y Máx Pooling
 * Filtros: 32
 * Tamaño del kernel: 3
 * padding: same
 * Función de Activación: relu

**Hyper parámetros de la capa de MaxPool:**

 * pool_size: 2 el tamaño es 2*2
 * strides: 2 el salto es de dos en dos
 * padding: valid. Cuando no se puede crear la matriz de max pooling 2*2  no se hace el borde

### 5. Capa de flattening
### 6. Capa de Fully connected
### 7. Capa de output

```python
model = tf.keras.Sequential()
# Primera capa
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=[32,32,3]))
# Segunda capa
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding='valid'))
# Tercera capa
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu'))
# Cuarta capa
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding='valid'))
# Quinta capa
model.add(tf.keras.layers.Flatten())
# Sexta capa
model.add(tf.keras.layers.Dense(units=128, activation='relu'))
# Septima capa
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
# Compilamos
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['sparse_categorical_accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=15, batch_size=512)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```
